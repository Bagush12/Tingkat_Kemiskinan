```python
# Fungsi untuk mengisi data yang hilang dengan rata-rata
def isi_data_hilang_dengan_rata_rata(df):
    for kolom in df.columns:
        if pd.api.types.is_numeric_dtype(df[kolom].dtype):  # Cek tipe data kolom
            rata_rata = df[kolom].mean()
            df[kolom].fillna(rata_rata, inplace=True)
```

```python
# Memanggil fungsi untuk mengisi data yang hilang dengan rata-rata
isi_data_hilang_dengan_rata_rata(df)

# Menampilkan DataFrame yang telah diisi
print(df)
```

# Metode Elbow untuk Tingkat Kemiskinan di Pulau Jawa

```python
# mengambil nilai dari kolom 'Jumlah penduduk miskin (000)
X = df[['Jumlah penduduk miskin (000)']].values

# membuat sebuah list (daftar) dengan nama wcss
wcss = []

# menghitung nilai Within-Cluster Sum of Squares (WCSS) untuk berbagai jumlah cluster (dari 1 hingga 10).
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# membuat sebuah grafik yang menampilkan hasil dari metode elbow (siku) dalam menentukan jumlah cluster yang optimal dalam analisis K-Means.
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS (Within-cluster sum of squares)')
plt.grid(True)
plt.show()
```

```python
# Mengambil subset dari DataFrame df yang terdiri dari kolom-kolom numerik dan menyimpannya dalam variabel X
X = df[numeric_columns]

# Melakukan penskalaan (scaling) data dalam variabel X menggunakan teknik StandardScaler dari pustaka scikit-learn (sklearn).
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# pengaturan nilai yang menentukan jumlah kluster yang akan digunakan dalam algoritma K-Means
n_clusters = 4

# Membuat objek SimpleImputer untuk menggantikan NaN dengan mean
imputer = SimpleImputer(strategy='mean')

# Mengisi nilai-nilai NaN dalam dataset dengan mean
X_imputed = imputer.fit_transform(X)
```

# Algortima Machine Learning Untuk Tingkat Kemiskinan di Pulau Jawa

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_imputed)

# mencetak atau menampilkan dua kolom dari DataFrame df, yaitu 'Kabupaten/Kota' dan 'Cluster'
print(df[['Kabupaten/Kota', 'Cluster']])
```

5.1 Melakukan clustering dengan K-Means

# Menghitung jarak data ke setiap pusat cluster

```python
# Menyimpan nama kota dalam DataFrame
kota = df['Kabupaten/Kota']

# Memilih fitur yang akan digunakan untuk clustering
X = df[['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']]

# Menghitung jarak data ke setiap pusat cluster
distances = pairwise_distances(X_scaled, kmeans.cluster_centers_)

# Membuat DataFrame dengan keterangan nama kota dan matriks jarak
distance_df = pd.DataFrame(distances, columns=[f'Cluster_{i+1}' for i in range(n_clusters)])
distance_df.insert(0, 'Kabupaten/Kota', kota)

# Menampilkan DataFrame dengan keterangan
print("Matrix Jarak dengan Nama Kota:")
print(distance_df)
```

```python
# Mengatur opsi untuk menampilkan semua baris
pd.set_option("display.max_rows", None) 
print("Matrix Jarak dengan Nama Kota:")
print(distance_df)
```

```python
# Menghitung jarak antara setiap data dengan pusat cluster terdekat
distances = kmeans.transform(X_scaled)  # Menggunakan transform untuk menghitung jarak

# Membuat DataFrame dengan keterangan nama kota dan jarak ke pusat cluster terdekat
distance_df = pd.DataFrame(distances, columns=[f'Jarak ke Cluster {i+1}' for i in range(n_clusters)])
distance_df.insert(0, 'Kabupaten/Kota', df['Kabupaten/Kota'])

#Menampilkan DataFrame dengan jarak ke pusat cluster terdekat
print("Jarak ke Pusat Cluster Terdekat:")
print(distance_df)
```

```python
# Range jumlah cluster yang akan diuji
range_clusters = range(2, 11)  # Ubah dari 1 menjadi 2 cluster

# Menyimpan Silhouette Score untuk setiap jumlah cluster
silhouette_scores = []

for n_clusters in range_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot grafik Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(range_clusters, silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.title('Metode Silhouette Score untuk Menentukan Jumlah Cluster Optimal')
plt.grid(True)
plt.show()
```
